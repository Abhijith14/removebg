```python
import pandas as pd
```

```python
df = pd.read_excel('mail.xlsx')
```

```python
df.columns
```

```python
# Step 1: Get unique student names
unique_students = df['Student Name'].unique()

# Create a new DataFrame to store the combined data
combined_df = pd.DataFrame({'Student Name': unique_students})

# Step 2 and 3: Iterate through the unique students and add column values
for student in unique_students:
    # Filter the original DataFrame for the current student
    student_df = df[df['Student Name'] == student]
    # Get the count of occurrences for the current student
    count = student_df.shape[0]
    
     # Iterate through the count and add column values to the combined DataFrame
    for i in range(count):
        clc_column = f'CLC{i+1}'
        stage_column = f'Stage{i+1}'
        cat_column = f'Cat{i+1}'

        # Add the remaing columns that are 'Class', 'Division', 'House Colour'
        combined_df.loc[combined_df['Student Name'] == student, 'Class'] = student_df.iloc[i]['Class']
        combined_df.loc[combined_df['Student Name'] == student, 'Division'] = student_df.iloc[i]['Division']
        combined_df.loc[combined_df['Student Name'] == student, 'House Colour'] = student_df.iloc[i]['House Colour']
        
        combined_df.loc[combined_df['Student Name'] == student, clc_column] = student_df.iloc[i]['Choose your Items in Cultural and Literary Competitions']
        combined_df.loc[combined_df['Student Name'] == student, stage_column] = student_df.iloc[i]['On-Stage/Off-Stage']
        combined_df.loc[combined_df['Student Name'] == student, cat_column] = student_df.iloc[i]['Cat']
```

```python
combined_df
```

```python
# Save the combined DataFrame to a new Excel file
combined_df.to_excel('output_file.xlsx', index=False)
```

```python
# change all cells having 'ii' to 'II' in column 'Cat1'
df['Cat1'] = df['Cat1'].str.upper()

# change all CLC1 to upper
df['CLC1'] = df['CLC1'].str.upper()
df['CLC2'] = df['CLC2'].str.upper()
df['CLC3'] = df['CLC3'].str.upper()
df['CLC4'] = df['CLC4'].str.upper()
df['CLC5'] = df['CLC5'].str.upper()
```

```python
# remove all the beginning and trailing spaces
df['CLC1'] = df['CLC1'].str.strip()
df['CLC2'] = df['CLC2'].str.strip()
df['CLC3'] = df['CLC3'].str.strip()
df['CLC4'] = df['CLC4'].str.strip()
df['CLC5'] = df['CLC5'].str.strip()
```

```python
# loop through all the values in column 'CLC1', 'CLC2', 'CLC3', 'CLC4', 'CLC5'

clcs = df[['CLC1', 'CLC2', 'CLC3', 'CLC4', 'CLC5']]

# get the unique values from clcs
unique_clcs = clcs.stack().unique()
len(unique_clcs)
```

```python
unique_clcs
```

```python
# go through each unique clc and get the count of occurrences in each column
final = {}
col1, col2, col3, col4 = [], [], [], []
for clc in unique_clcs:
    # print(clc)

    c1 = df[df['CLC1'] == clc]['Cat1'].value_counts().to_dict()
    c2 = df[df['CLC2'] == clc]['Cat1'].value_counts().to_dict()
    c3 = df[df['CLC3'] == clc]['Cat1'].value_counts().to_dict()
    c4 = df[df['CLC4'] == clc]['Cat1'].value_counts().to_dict()
    c5 = df[df['CLC5'] == clc]['Cat1'].value_counts().to_dict()

    # print(c1)
    # print(c2)
    # print(c3)
    # print(c4)
    # print(c5)

    # combine and merge all the dictionaries while adding the values
    main_dict = {k: c1.get(k, 0) + c2.get(k, 0) + c3.get(k, 0) + c4.get(k, 0) + c5.get(k, 0) for k in set(c1) | set(c2) | set(c3) | set(c4) | set(c5)}
    # print(main_dict)

    col1.append(clc)
    col2.append(main_dict.get('I', 0))
    col3.append(main_dict.get('II', 0))
    col4.append(main_dict.get('III', 0))
    # print()

final['CLC'] = col1
final['Cat-1'] = col2
final['Cat-2'] = col3
final['Cat-3'] = col4
```

```python
final = pd.DataFrame(final)

# create a new column 'Total' and add the sum of all the columns
final['Total'] = final['Cat-1'] + final['Cat-2'] + final['Cat-3']

final
```

```python
# save the final DataFrame to a new Excel file
final.to_excel('final.xlsx', index=False)
```

```python
import pandas as pd

# Assuming your dataframe is named 'df'
# Create a new empty DataFrame to store the transformed data
df_transformed = pd.DataFrame(columns=['Student Name', 'Class', 'Division', 'House Colour', 'Cat1', 'CLC', 'Stage'])

# Iterate through each row of the original DataFrame
for _, row in df.iterrows():
    student_info = row[['Student Name', 'Class', 'Division', 'House Colour', 'Cat1']].copy()
    
    # Iterate through each event (Cat1, CLC, Stage) and create a new row for each
    for i in range(1, 6):
        # cat = row['Cat{}'.format(i)]
        clc = row['CLC{}'.format(i)]
        stage = row['Stage{}'.format(i)]
        
        # Check if both 'CLC' and 'Stage' are not NaN before creating a new row
        if pd.notna(clc) and pd.notna(stage):
            new_row = student_info.copy()
            # new_row['Cat1'] = cat
            new_row['CLC'] = clc
            new_row['Stage'] = stage
            df_transformed = df_transformed.append(new_row, ignore_index=True)

# Optionally, sort the dataframe if needed
df_transformed = df_transformed.sort_values(by=['Student Name', 'Class', 'Division', 'House Colour']).reset_index(drop=True)

# Display the transformed dataframe
df_transformed
```

```python
# save to a new Excel file
df_transformed.to_excel('transformed.xlsx', index=False)
```
